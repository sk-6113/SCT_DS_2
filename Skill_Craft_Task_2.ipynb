```python
!pip install yfinance pandas matplotlib seaborn

import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

amazon = yf.download('AMZN', start='2010-01-01', end='2025-01-01')

print("Amazon Stock Price Data (First 5 Rows):")
print(amazon.head())

print("\nData Info:")
amazon.info()

print("\nSummary Statistics:")
print(amazon.describe())

plt.figure(figsize=(12, 6))
plt.plot(amazon['Close'], label='Closing Price', color='blue')
plt.title('Amazon (AMZN) Stock Closing Price Over Time')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.legend()
plt.grid(True)
plt.show()

amazon['MA50'] = amazon['Close'].rolling(window=50).mean()
amazon['MA200'] = amazon['Close'].rolling(window=200).mean()

plt.figure(figsize=(14, 7))
plt.plot(amazon['Close'], label='Closing Price', alpha=0.7)
plt.plot(amazon['MA50'], label='50-Day MA', linestyle='--')
plt.plot(amazon['MA200'], label='200-Day MA', linestyle='--')
plt.title('Amazon Stock Price with 50 & 200 Day Moving Averages')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.legend()
plt.grid(True)
plt.show()

plt.figure(figsize=(12, 5))
plt.plot(amazon['Volume'], color='purple')
plt.title('Amazon Trading Volume Over Time')
plt.xlabel('Date')
plt.ylabel('Volume')
plt.grid(True)
plt.show()

plt.figure(figsize=(8, 5))
sns.heatmap(amazon.corr(), annot=True, cmap='YlGnBu')
plt.title('Correlation Matrix - Amazon Stock Data')
plt.show()
```
